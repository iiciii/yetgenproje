```python
import pandas as pd
import numpy as np
import time 
import random 

df = pd.read_excel("Tarot kart anlamları 2.csv")

class tarotcu_abla():

    def __init__(self, df):
        self.df = df

    def rastgeleCek(self,kartSayisi):
        return self.df.sample(n = kartSayisi)

    def tekKart():
        print("Bir soru düşün ve kartın dediklerine güven!")
        print("Karıştırılıyor ve senin için bir kart çekiliyor...")
        time.sleep(3)
        #excelden rastgele hücre seçimi ve ekrana yazdırma


    def ücKart(): 
        i = 0
        while i < 3:
            if i == 0:
               #  excelden geçmiş kartı çeiklir.
                print("Geçmiş kartı: ?????? ")
                i += 1
            if i == 1:
                # excelden şimdi kartı çeiklir.
                print("Şimdi kartı:  ?????")
                i += 1
            else:
                #  excelden gelecek kartı çeiklir.
                print("gelecek kartı:  ?????")
                i += 1

    def altıKart():
        i = 0
        while i < 6:
            if i == 0:
               #  excelden geçmiş kartı çeiklir.
                print("kendisi kartı: ?????? ")
                i += 1
            if i == 1:
                # excelden şimdi kartı çeiklir.
                print("aile kartı:  ?????")
                i += 1
            if i == 2:
                #  excelden gelecek kartı çeiklir.
                print("umdukları kartı:  ?????")
                i += 1
            if i == 3:
               #  excelden geçmiş kartı çeiklir.
                print("arkadaşları kartı: ?????? ")
                i += 1
            if i == 4:
                # excelden şimdi kartı çeiklir.
                print("korkuları kartı:  ?????")
                i += 1
            else:
                #  excelden gelecek kartı çeiklir.
                print("??? kartı:  ?????")
                i += 1



tarotcu_abla = tarotcu_abla()

print("""
TAROTÇU ABLA'YA HOŞ GELDİNİZ?
      
      1.Tek kart ---> kısa cevaplı sorular
      2.Üç kart ---> geçmiş/şimdi/gelecek
      3.Altı kart ---> kendi/arkadaşları/aile/umdukları/korkuların/???

      **çıkmak için 'q' ya basınız

      
      """)


while True:
    girdi = input("Çekilecek kart sayısını seçiniz")
    if (girdi.lower == "q"):
        print("Fal baktırmadığınız için üzgünüz.")
        break
    elif (girdi == "1"):
        tarotcu_abla.tekKart()
    elif (girdi == "2"):
        tarotcu_abla.ücKart()
    elif (girdi == "3"):
        tarotcu_abla.altıKart()
    else:
        print("Tarotçu Abla yanlış bir seçim yaptığını hissetti!!")
   
```
